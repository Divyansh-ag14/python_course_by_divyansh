# File Handling: File Operations

## Introduction
Learn how to read from and write to files in Python.

## Topics Covered:
1. Opening and Closing Files
2. Reading Files
3. Writing Files
4. File Modes
5. Context Managers (with statement)


```python
# Writing to a file
with open("example.txt", "w") as file:
    file.write("Hello, Python!\n")
    file.write("This is a test file.\n")
    file.write("File handling is important!")

print("File written successfully!")

# Reading from a file
with open("example.txt", "r") as file:
    content = file.read()
    print("\nFile content:")
    print(content)
```

## Different Reading Methods


```python
# Create a sample file
with open("sample.txt", "w") as f:
    f.write("Line 1\nLine 2\nLine 3\nLine 4\nLine 5")

# read() - reads entire file
with open("sample.txt", "r") as f:
    content = f.read()
    print("Using read():")
    print(content)

# readline() - reads one line at a time
print("\nUsing readline():")
with open("sample.txt", "r") as f:
    line = f.readline()
    while line:
        print(line.strip())
        line = f.readline()

# readlines() - reads all lines into a list
print("\nUsing readlines():")
with open("sample.txt", "r") as f:
    lines = f.readlines()
    for line in lines:
        print(line.strip())

# Iterating over file object
print("\nIterating over file:")
with open("sample.txt", "r") as f:
    for line in f:
        print(line.strip())
```
