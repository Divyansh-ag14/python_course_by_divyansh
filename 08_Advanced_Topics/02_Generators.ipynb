# Advanced Topics: Generators

## Introduction
Generators are a simple way to create iterators using functions.

## Topics Covered:
1. Generator Functions
2. Generator Expressions
3. yield vs return
4. Generator Methods
5. When to Use Generators


```python
# Generator function
def countdown(n):
    while n > 0:
        yield n
        n -= 1

# Using generator
for num in countdown(5):
    print(num)

# Generator expression
squares = (x**2 for x in range(10))
print(f"\nGenerator object: {squares}")
print(f"First few values: {[next(squares) for _ in range(5)]}")
```
